```python
def getUrlContent(url):
    cleanUrl = request.urlopen(url).read().decode('utf-8')
    cleanUrl = re.sub(r'\n', ' ', cleanUrl)
    cleanUrl = re.sub(r'<script.*?script>', '', cleanUrl, re.S)
    cleanUrl = re.sub(r'<.*?>', '', cleanUrl)
    cleanUrl = re.sub(r'\s+', ' ', cleanUrl)
    return cleanUrl

print (request.urlopen('http://www.nltk.org/book/ch03.html').read().decode('utf-8')[:40])
print (getUrlContent('http://www.nltk.org/book/ch03.html')[:40])
```
